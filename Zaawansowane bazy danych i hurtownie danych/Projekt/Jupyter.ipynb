```python
# Korelacja miÄ™dzy zmiennymi przedstawiona na wykresie heatmapy
# bliÅ¼ej zÃ³Å‚tego koloru oznacza wysokÄ… dodatniÄ… korelacje, bliÅ¼ej filoetowego koloru wysokÄ… ujemnÄ… korelacje
# korelacja dodatnia jest wtedy gdy jedna zmienna roÅ›nie to druga takÅ¼e
# korelacja ujemna jest wtedy gdy jedna zmienna roÅ›nie druga zmienna maleje
# korelacja bliska zeru nie ma praktycznie znaczenia 
df_copy = df.copy().apply(lambda x: x.factorize()[0]).corr()
print("Zmienne: \n")
for i, col in enumerate(list(df_copy.columns)):
    print(f"{i}: {col}")
plt.matshow(df.corr())
plt.colorbar()
plt.show()
```

```python
# Grupowanie danych wzglÄ™dem zmiennych kategoralnych + dane statystyczne (Å›rednia, kwantyle (25%, 75%), mediana (50%) i ilosÄ‡ danych branych pod uwagÄ™).

non_categorical = [
    'BMI',
    'MentHlth',
    'PhysHlth',
]

rest_cols = list(set(df.columns) - set(non_categorical))
data = []

for i, var in enumerate(rest_cols):
    rest_cols_without_var = list(set(df.columns) - set([var]))
    grouped_df = df.groupby(var)
    selected = grouped_df[rest_cols_without_var].describe()
    data.append(selected)
    
    display(HTML(f"<h1>Zmienna: {var} </h1>"))
    for col in rest_cols_without_var:
        display(HTML(f"<h3> {col} <h3>"))
        col_stats = selected[col][['mean', '25%', '50%', '75%', 'count']] # brak brakujÄ…cych danych, plik jest uzupeÅ‚niony w 100%
        display(col_stats)
```
